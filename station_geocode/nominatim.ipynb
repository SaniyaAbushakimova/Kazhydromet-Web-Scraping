```python
from geopy.geocoders import Nominatim
import pandas as pd
```

```python
df = pd.read_csv('region_station_list.csv')
df.rename(columns={"stn": "station"}, inplace = True)
```

```python
region_names = {'KZ-AKM': 'Акмолинская область',
                'KZ-AKT': 'Актюбинская область',
                'KZ-ALM': 'Алматинская область', 
                'KZ-ATY': 'Атырауская область', 
                'KZ-KAR': 'Карагандинская область', 
                'KZ-KUS': 'Костанайская область',
                'KZ-KZY': 'Кызылординская область', 
                'KZ-MAN': 'Мангистауская область', 
                'KZ-PAV': 'Павлодарская область', 
                'KZ-SEV': 'Северо-Казахстанская область', 
                'KZ-VOS': 'Восточно-Казахстанская область', 
                'KZ-YUZ': 'Туркестанская область',
                'KZ-ZAP': 'Западно-Казахстанская область', 
                'KZ-ZHA': 'Жамбылская область'
                }

temp = df['region']
df = df.apply(lambda x: x.replace(region_names, regex=True))
df['region_code'] = temp
df['location'] = df['station']+ ', ' + df['region'] + ', ' + 'Казахстан'
```

```python
df.head()
```

```python
# Find station coordinates with Nominatim
nom = Nominatim(user_agent='kazhydromet')
df['coordinates'] = df['location'].apply(nom.geocode, timeout=10)
df['latitude'] = df['coordinates'].apply(lambda x: x.latitude if x!= None else None)
df['longitude'] = df['coordinates'].apply(lambda x: x.longitude if x!= None else None)
```

```python
# Stations that were not identified by Nominatim 
df[df['latitude'].isnull()==True]
```

```python
# Manually searching for the geocodes of stations that were not identified by Nominatim
additional = pd.read_csv('additional.csv')
additional.head()
```

```python
# Add the above stations to the main list of coordinates
for index, row in additional.iterrows():
    df.loc[df['station']==row['station'],'latitude'] = row['latitude']
    df.loc[df['station']==row['station'],'longitude'] = row['longitude']
```

```python
# Stations that could not be found manually
list(df['station'][df['latitude'].isnull()==True])
```

```python
# Manually cleaned coordinates file
additional_2 = pd.read_csv('additional_2.csv')
additional_2.head()
```

```python
for index, row in additional_2.iterrows():
    df.loc[df['station']==row['station'],'latitude'] = row['latitude']
    df.loc[df['station']==row['station'],'longitude'] = row['longitude']
```

```python
# Save final csv with station coordinates
df.to_csv('final_coordinates.csv')
```
